```python
import flwr as fl
import torch
from torch.utils.data import DataLoader, TensorDataset

# --- 1. THE ML MODEL & TRAINING LOOP ---
class MedicalModel(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.fc = torch.nn.Linear(10, 2) # Example: Binary classification

    def forward(self, x): return self.fc(x)

def train_local(model, loader, device):
    optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
    model.train()
    for images, labels in loader:
        # MOVE DATA TO GPU HERE
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        loss = torch.nn.CrossEntropyLoss()(model(images), labels)
        loss.backward()
        optimizer.step()

# --- 2. THE HOSPITAL CLIENT (Node) ---
class HospitalClient(fl.client.NumPyClient):
    def __init__(self, hospital_id, data_loader):
        self.id = hospital_id
        self.loader = data_loader
        # Determine device
        self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        self.model = MedicalModel().to(self.device)

    def fit(self, parameters, config):
        # A. Set Global Weights
        for p, git_p in zip(self.model.parameters(), parameters):
            p.data = torch.from_numpy(git_p).to(self.device) # Ensure weights are on GPU

        # B. Train (Passing the device)
        train_local(self.model, self.loader, self.device)

        print(f"Hospital {self.id} notarized update on Sepolia.")

        # Return weights back to CPU for Flower aggregation
        return [val.cpu().detach().numpy() for val in self.model.parameters()], len(self.loader.dataset), {}

# --- 3. THE SIMULATION LAUNCHER ---
def client_fn(cid):
    dummy_data = DataLoader(TensorDataset(torch.randn(100, 10), torch.randint(0, 2, (100,))), batch_size=10)
    # Convert NumPyClient to the standard Client expected by the new API
    return HospitalClient(cid, dummy_data).to_client()

if __name__ == "__main__":
    fl.simulation.start_simulation(
        client_fn=client_fn,
        num_clients=10,
        client_resources={"num_gpus": 0.2}, # 5 hospitals parallel per GPU
        config=fl.server.ServerConfig(num_rounds=3),
    )
```

```python
import flwr as fl
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset, random_split
import medmnist
from medmnist import INFO

# --- 1. SETUP REAL MEDICAL DATA (PneumoniaMNIST) ---
def load_partitioned_data(num_clients=10):
    # Download Data info
    data_flag = 'pneumoniamnist'
    info = INFO[data_flag]
    DataClass = getattr(medmnist, info['python_class'])

    # Load the training set
    train_dataset = DataClass(split='train', download=True, transform=None)

    # Convert to standard Tensor format (1 channel, 28x28 images)
    # MedMNIST images are small (28x28) just like MNIST
    X = torch.tensor(train_dataset.imgs).float().unsqueeze(1) / 255.0 # Normalize 0-1
    y = torch.tensor(train_dataset.labels).long().squeeze()

    full_dataset = TensorDataset(X, y)

    # Split into 10 "Hospitals"
    partition_size = len(full_dataset) // num_clients
    lengths = [partition_size] * num_clients
    # Handle any remainder
    if sum(lengths) < len(full_dataset):
        lengths[0] += len(full_dataset) - sum(lengths)

    datasets = random_split(full_dataset, lengths)
    return datasets

# Global variable to hold partitions so we don't reload every time
HOSPITAL_DATASETS = load_partitioned_data()

# --- 2. DEFINE A REALISTIC CNN MODEL ---
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        # Simple CNN for 28x28 Greyscale X-rays
        self.conv1 = nn.Conv2d(1, 32, kernel_size=3)
        self.relu = nn.ReLU()
        self.pool = nn.MaxPool2d(2)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3)
        self.fc1 = nn.Linear(64 * 5 * 5, 128)
        self.fc2 = nn.Linear(128, 2) # Binary: Pneumonia vs Normal

    def forward(self, x):
        x = self.pool(self.relu(self.conv1(x)))
        x = self.pool(self.relu(self.conv2(x)))
        x = x.view(-1, 64 * 5 * 5)
        x = self.relu(self.fc1(x))
        x = self.fc2(x)
        return x

def train_local(model, loader, device, epochs=1):
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=0.001)
    model.train()
    for _ in range(epochs):
        for images, labels in loader:
            images, labels = images.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

# --- 3. THE FLOWER CLIENT ---
class HospitalClient(fl.client.NumPyClient):
    def __init__(self, hospital_id, dataset):
        self.id = hospital_id
        # Create a local DataLoader for this hospital
        self.loader = DataLoader(dataset, batch_size=32, shuffle=True)
        self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        self.model = Net().to(self.device)

    def fit(self, parameters, config):
        # Update local model with global weights
        self.set_parameters(parameters)

        # Train on local X-ray data
        train_local(self.model, self.loader, self.device, epochs=1)

        # (Simulated) Blockchain Notarization
        print(f"Hospital {self.id}: Trained on {len(self.loader.dataset)} X-rays. Hash sent to Sepolia.")

        return self.get_parameters(config={}), len(self.loader.dataset), {}

    def set_parameters(self, parameters):
        for p, git_p in zip(self.model.parameters(), parameters):
            p.data = torch.from_numpy(git_p).to(self.device)

    # --- FIX IS HERE: Added 'config' argument ---
    def get_parameters(self, config):
        return [val.cpu().detach().numpy() for val in self.model.parameters()]

# --- 4. SIMULATION ---
def client_fn(cid):
    # Map the Client ID (string "0"-"9") to the dataset partition
    hospital_idx = int(cid)
    hospital_data = HOSPITAL_DATASETS[hospital_idx]
    return HospitalClient(cid, hospital_data).to_client()

if __name__ == "__main__":
    fl.simulation.start_simulation(
        client_fn=client_fn,
        num_clients=10,
        client_resources={"num_gpus": 0.1}, # T4 is powerful enough for 10 small CNNs
        config=fl.server.ServerConfig(num_rounds=5),
    )
```

```python
import flwr as fl
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset, random_split
import medmnist
from medmnist import INFO

# --- 1. SETUP REAL MEDICAL DATA (PneumoniaMNIST) ---
def load_partitioned_data(num_clients=10):
    # Download Data info
    data_flag = 'pneumoniamnist'
    info = INFO[data_flag]
    DataClass = getattr(medmnist, info['python_class'])

    # Load the training set
    train_dataset = DataClass(split='train', download=True, transform=None)

    # Convert to standard Tensor format (1 channel, 28x28 images)
    # MedMNIST images are small (28x28) just like MNIST
    X = torch.tensor(train_dataset.imgs).float().unsqueeze(1) / 255.0 # Normalize 0-1
    y = torch.tensor(train_dataset.labels).long().squeeze()

    full_dataset = TensorDataset(X, y)

    # Split into 10 "Hospitals"
    partition_size = len(full_dataset) // num_clients
    lengths = [partition_size] * num_clients
    # Handle any remainder
    if sum(lengths) < len(full_dataset):
        lengths[0] += len(full_dataset) - sum(lengths)

    datasets = random_split(full_dataset, lengths)
    return datasets

# Global variable to hold partitions so we don't reload every time
HOSPITAL_DATASETS = load_partitioned_data()

# --- 2. DEFINE A REALISTIC CNN MODEL ---
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        # Simple CNN for 28x28 Greyscale X-rays
        self.conv1 = nn.Conv2d(1, 32, kernel_size=3)
        self.relu = nn.ReLU()
        self.pool = nn.MaxPool2d(2)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3)
        self.fc1 = nn.Linear(64 * 5 * 5, 128)
        self.fc2 = nn.Linear(128, 2) # Binary: Pneumonia vs Normal

    def forward(self, x):
        x = self.pool(self.relu(self.conv1(x)))
        x = self.pool(self.relu(self.conv2(x)))
        x = x.view(-1, 64 * 5 * 5)
        x = self.relu(self.fc1(x))
        x = self.fc2(x)
        return x

# --- UPDATED TRAINING FUNCTION ---
def train_local(model, loader, device, name="Client", epochs=1):
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=0.001)
    model.train()

    for epoch in range(epochs):
        running_loss = 0.0
        correct = 0
        total = 0

        for images, labels in loader:
            images, labels = images.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            # Track metrics
            running_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

        avg_loss = running_loss / len(loader)
        accuracy = 100 * correct / total

        # This print statement is what you were missing:
        print(f"🏥 {name} | Epoch {epoch+1} | Loss: {avg_loss:.4f} | Accuracy: {accuracy:.2f}%")

# --- UPDATED CLIENT ---
class HospitalClient(fl.client.NumPyClient):
    def __init__(self, hospital_id, dataset):
        self.id = hospital_id
        self.loader = DataLoader(dataset, batch_size=32, shuffle=True)
        self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        self.model = Net().to(self.device)

    def fit(self, parameters, config):
        self.set_parameters(parameters)

        # Pass the Hospital ID to the training function
        train_local(self.model, self.loader, self.device, name=f"Hospital {self.id}", epochs=1)

        print(f"✅ Hospital {self.id} finished training. Hash sent to Sepolia.")
        return self.get_parameters(config={}), len(self.loader.dataset), {}

    def set_parameters(self, parameters):
        for p, git_p in zip(self.model.parameters(), parameters):
            p.data = torch.from_numpy(git_p).to(self.device)

    def get_parameters(self, config):
        return [val.cpu().detach().numpy() for val in self.model.parameters()]

# --- 4. SIMULATION ---
def client_fn(cid):
    # Map the Client ID (string "0"-"9") to the dataset partition
    hospital_idx = int(cid)
    hospital_data = HOSPITAL_DATASETS[hospital_idx]
    return HospitalClient(cid, hospital_data).to_client()

if __name__ == "__main__":
    fl.simulation.start_simulation(
        client_fn=client_fn,
        num_clients=10,
        client_resources={"num_gpus": 0.1}, # T4 is powerful enough for 10 small CNNs
        config=fl.server.ServerConfig(num_rounds=5),
    )
```
